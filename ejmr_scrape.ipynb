```python
from bs4 import BeautifulSoup
import csv
import urllib2
import datetime
```

```python
def ejmr_scrape(pg_num, csv_filename):
    """
    Columns
        title: post title including link to post
        num_posts: number of posts as of download date
        views: number of views as of download date
        votes: upvote-downvote
        freshness: time elapsed since last post
        download_time: date and time when data was scrapped (EST)
    """
#     print "{0}".format(pg_num)

    if pg_num % 10 == 0:
        
        print "Webpage number: {0}".format(pg_num)
    
    page = urllib2.urlopen("https://www.econjobrumors.com/page/{0}".format(pg_num))
    soup = BeautifulSoup(page)
    
    # print "Open link and pass into BeautifulSoup..."

    table = soup.find("table", id="latest").find_all("tr") # table containing posts
    
    # print table

    ejmr_filename = "ejmr_data/{0}{1}.csv".format(csv_filename, pg_num)
    
    header_row = ["title",
                  "num_posts",
                  "views",
                  "votes",
                  "freshness",
                  "download_time"
                 ]

    # print "Begin collection of files..."
    
    with open(ejmr_filename, "wb") as f:

        writeFile = csv.writer(f, quoting=csv.QUOTE_NONNUMERIC)

        writeFile.writerow(header_row)

        for tb in table:

            # get the link to the first page of a thread
            try:
        
                tbfindall = tb.find_all("td")
                link =  tbfindall[0].find_all('a')  # get the part containing <a href = ....>
                link_in_href = link[0].get('href')  # get the actual link as string
                print link_in_href, type(link_in_href)
            
            except Exception as e:
                
                print e


            # !!! NEXT use link_in_href to get first post in thread !!!
            
            
                
#             row = []

#             for elem in tb.find_all("td"):

#                 # remove empty element from list
#                 elem_list = [e.encode('utf-8').strip() for e in elem]
                
# #                 print elem_list
                
#                 try:
                    
#                     elem_list = filter(None, elem_list)

#                 except:

#                     print "no blank"

#                 # get only first element of list (addresses case where title includes page number links next to them---e.g. Top theorists graduating next year? - 23…8910)
#                 elem_list_first = elem_list[0]
                
#                 print elem_list
#                 print elem_list_first
                
#                 row.append(elem_list_first)

# #             if row:  # write row only if non-empty
# #                 row.append(datetime.datetime.now())
# #                 writeFile.writerow(row)


            print "=============="
```
